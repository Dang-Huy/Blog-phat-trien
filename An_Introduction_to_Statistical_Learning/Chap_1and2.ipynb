# Oke, đầu tiên Statistical Learning là gì ?
## Statistical Learning chính là lý thuyết nền tảng giúp hiểu và phân tich các model Machine Learning từ góc nhìn toán học.  
- SL tập trung hơn vào hiểu, diễn giải model trong khi ML tập trung vào hiệu suất dự đoán của model. 
- SL chủ yếu tập trung vào regression và classification trong khi ML bao gồm SL, deep learning, reinforcement. 

## Như các bạn đã học trong Statistics thì phương trình ở ảnh đầu là mô hình tổng thể (population model) còn ở ảnh 2 là mô hình mẫu (sample model). 
## Mục dích của chúng ta là ước lượng f. Nhưng tại sao phải làm vậy ?

## Lý do đầu tiên là để dự đoán được Y (Prediction)
### Nhưng chúng ta luôn luôn không thể tạo ra được f đủ tốt để dự đoán chính xác Y. Trong rất nhiều context khác nhau, chúng ta có thể có toàn bộ dữ liệu input (nghĩa là chúng ta có X thay vì chỉ là $\hat{X}$). Nhưng để đưa ra được output chính xác là điều hoàn toàn không thể vì population model của Y luôn gồm error term. 
- Phần error này bao gồm phần có thể giảm bớt được (reduccible error) (do X chưa đủ lớn, do f chưa đủ chính xác, ...) và phần không thể giảm bớt (irreduccible error) do đây là phần sai số đến từ yếu tố ngẫu nhiên
- Do đó chúng ta có thể ước lượng f bằng sample model <br>

- Which predictors are associated with the response
- What is the relationship between the response and each predictor?
- Can the relationship between Y and each predictor be adequately sum
marized using a linear equation, or is the relationship more compli
cated? 
### Sẽ có những thuật toán có khả năng predict nhưng không có khả năng inference tốt và ngược lại và cũng có những algo làm được cả 2
### Đối với các thuật toán thuộc Parameter method (Ex: Linear Regression, Naive Bayes Classifier) thì khả năng Inference tốt hơn Non-para method (Ex: Support Vector Machine, K-Nearest Neighbour)
